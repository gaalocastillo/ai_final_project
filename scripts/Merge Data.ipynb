### Education data

```python
df_education1 = pd.read_excel('../data/education_extracted1.xlsx')
df_education1['diario'] = 'El Comercio'
df_education2 = pd.read_excel('../data/education_extracted2.xlsx')
df_education1['diario'] = 'El Universo'
frames = [df_education1, df_education2]
df_education = pd.concat(frames)
```

```python
df_education.head()
```

```python
df_education['svm_prediction'] = 'educ'
df_education['year'] = None
df_education['id_news'] = None
df_education = df_education.drop(columns=['link'], inplace=False)
```

```python
df_education.to_csv('../data/education.csv', sep='|', index=False, encoding='utf8')
```
